```python
import nltk
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv('nlp_data.csv')
```

```python
data.columns = ['Text', 'Target']
data.head()
```

```python
#Remove stopwords from column 'Text'

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

#Converting Text column of dataframe into an array
X = np.array(data.Text)


#Pre-Process Feature Extraction
################################
################################
################################

#Sentence Count
from nltk.tokenize import sent_tokenize
sent_count = []
for i in range(0,len(X)):
    sent = sent_tokenize(X[i])
    sent_count.append(len(sent))
    
#Word Count in each Sentence
############################

from nltk.tokenize import word_tokenize
word_count = []
for i in range(0,len(X)):
    word = word_tokenize(X[i])
    word_count.append(len(word))


#Feature Extraction for Tagged Words
###################################

Determiner = []
Noun = []
ForeignWord = []
Adjective = []
Preposition = []
Verb = []
Adverb = []
Pronoun = []
Interjection = []
Others = []
    
for i in range(0,len(X)):
    
    DeterminerCount = 0
    NounCount = 0
    ForeignWordCount = 0
    AdjectiveCount = 0
    PrepositionCount = 0
    VerbCount = 0
    AdverbCount = 0
    PronounCount = 0
    InterjectionCount = 0
    OthersCount = 0
    
    word_token = nltk.word_tokenize(X[i])
    tagged_word = nltk.pos_tag(word_token)    
    
    em = []
    for w in tagged_word:
        em.append(w[1])
    
    for w in range(0,len(em)):
    
        if em[w]  == "NN" or em[w]  == "NNS" or em[w]  == "NNP" or em[w]  == "NNPS" :
            NounCount += 1
        elif em[w]  == "VB" or em[w]  == "VBD" or em[w]  == "VBG" or em[w]  == "VBN" or em[w]  == "VBP" or em[w]  == "VBZ" :
            VerbCount += 1
        elif em[w]  == "RB" or em[w]  == "RBR" or em[w]  == "RBS" :
            AdverbCount += 1
        elif em[w]  == "JJ" or em[w]  == "JJR" or em[w]  == "JJS" :
            AdjectiveCount += 1
        elif em[w]  == "PRP" or em[w]  == "PRP$" :
            PronounCount += 1
        elif em[w]  == "IN" :
            PrepositionCount += 1
        elif em[w]  == "DT" :
            DeterminerCount += 1  
        elif em[w]  == "UH" :
            InterjectionCount += 1 
        elif em[w]  == "FW" :
            ForeignWordCount += 1 
        else:
            OthersCount +1
    
    Determiner.append(DeterminerCount)
    Noun.append(NounCount)
    ForeignWord.append(ForeignWordCount)
    Adjective.append(AdjectiveCount)
    Preposition.append(PrepositionCount)
    Verb.append(VerbCount)
    Adverb.append(AdverbCount)
    Pronoun.append(PronounCount)
    Interjection.append(InterjectionCount)
    Others.append(OthersCount)

################################
################################
################################

#Defining Stop Words
sw = list(stopwords.words('english'))
custom_sw = [] #CUSTOM INPUT OF STOPWORDS
for word in custom_sw:
    sw.append(word)

#Regular Expressions, removing non-english characters and changing all to lower case
corpus = []
for i in range(0,len(X)):
    review = re.sub(r'\W', ' ', str(X[i]))
    review = review.lower()
    review = re.sub(r'\s+[a-z]\s+', ' ', review)
    review = re.sub(r'^[a-z]\s+', ' ', review)
    review = re.sub(r'\s+', ' ', review)
    corpus.append(review)
    
#Removing stop words
filtered_corpus = []
for i in range(0,len(corpus)):    
    filtered = [w for w in word_tokenize(corpus[i]) if w not in sw]
    filtered = ' '.join(filtered)
    filtered_corpus.append(filtered)
    
#Lemmatization
from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet') DEPENDENCIES
lm = WordNetLemmatizer()
corpus_stemmed = []
for i in range(0,len(filtered_corpus)):
    words = []
    word_token = word_tokenize(filtered_corpus[i])
    for w in range(0,len(word_token)):
        words.append(lm.lemmatize(word_token[w]))
        words_ = ' '.join(words)
    corpus_stemmed.append(words_)
```
